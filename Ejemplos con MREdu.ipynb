# Distribución del numero de palabras por linea

Vamos a hacer una pequeña prueba. Para ello tomaremos el texto de un libro y vamos a determinar cual es la distribución de palabras por linea en dicho texto.

Empezaremos cargando algunas funciones que vamos a necesitar:

```python
from mredu.simul import map_red, input_file, run
from re import split
```

El primer paso es definir el mapper de nuestro proceso:

```python
def mymap(k, v):
    words = list(filter(lambda s: s != '', split(r'\W', v)))
    return len(words), 1
```

Este se limita a tomar de cada par el valor (v), que es el que contiene en nuestro ejempo el texto de una linea, y dividirlo en palabras generando unos nuevos pares (numero_palabras, 1). En nuestro caso el 1 nos servirá para posteriormente sumarlo y determinar cuantas veces hemos visto ese número de palabras.

Veamos cual es la salida de nuestro ejemplo:

```python
map_only = map_red(input_file('data/quijote.txt'), mapper=mymap)
run(map_only)
```

Para obtener el resultado final necesitamos sumar los diferentes valores asociados a las claves (que son las distintas longitudes de linea).

Nestro reducer es:

```python
def myred(k, vs):
    return k, sum(vs)
```

Ahora ejecutemos nuestro proceso completo:

```python
process = map_red(input_file('data/quijote.txt'),
                  mapper=mymap,
                  reducer=myred)
run(process)
```

# Contaje de palabras

Usando el mismo texto podemos ver como contar cuantas veces aparece cada palabra individual, quitando las palabras que se usan para conectar las frases y dejando las que aportan significado.

Para ello necesitamos un nuevo mapper:

```python
stopwords = ['así', 'para', 'sus', 'una', 'ni', 'porque', 'sin', 'tan', 'al',
             'si', 'me', 'un', 'más', 'es', 'del', 'lo', 'las', 'le', 'mas',
             'por', 'su', 'con', 'los', 'se', 'no', 'en', 'el', 'la', 'a', 'y',
             'de', 'que', 'muy', 'qué', 'como', 'mi', 'o', 'aquel', 'ya',
             'pues', 'cuando', 'cual', 'pero', 'este', 'esto', 'aquí',
             'aquella', 'aquello', 'él', 'ha', 'entre', 'd']

def mymap(_, v):
    words = list(filter(lambda s: s != '', split(r'\W', v)))
    return [(word.lower(), 1) for word in words if word.lower() not in stopwords]
```

Es similar al anterior pero devuelve pares (palabra, 1) y solo lo hace para aquellas palabras que no están en *stop_words*.

Veamoslo en acción:

```python
map_only = map_red(input_file('data/quijote.txt'), mapper=mymap)
run(map_only)
```

Por último usemos un reducer para generar el conaje final:

```python
def myred(k, vs):
    total = sum(vs)
    if total >= 100:
        return k, sum(vs)
    else:
        return None
```

Todo junto nos da este resultado:

```python
process = map_red(input_file('data/quijote.txt'),
                  mapper=mymap,
                  reducer=myred)
run(process)
```
